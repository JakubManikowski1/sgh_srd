```python
%%time

import random
import math
import datetime
import pandas as pd


pd.set_option('display.float_format', lambda x: '%.3f' % x)

df = pd.DataFrame(columns=['Liczba_symulacji','Pole_kola','Dopasowanie'])


def main(start=100_000, step=50_000, stop=10_000_000, r=0.5):

    print("Stała PI = ", math.pi, ".\n", sep = "")
    print(f"Pole koła o r = {r}, równa się = ")
    print(f'= {(math.pi*r*r):.8f}', ".", sep = "")
    
    
    timeStart = datetime.datetime.now()
    print("Start:", timeStart)
    print("Please wait")
    
    for i in range(start,stop+1,step):
        liczba_punktow = i
        print("Bieżąca liczba symulacji:", i)
        licznik = 0

        for j in range(1,liczba_punktow,1):
            x = random.uniform(0,1)
            y = random.uniform(0,1)

            if x*x + y*y <= 1:
                licznik += 1
        
        pi = 4.0 * licznik / liczba_punktow
        pole_kola = pi * r*r
        pole_kola_real = math.pi * r * r
        dopasowanie = 1 - (abs(pole_kola_real - pole_kola)  / pole_kola_real )
        df.loc[len(df)] = [liczba_punktow, pole_kola, dopasowanie]
        
    timeStop = datetime.datetime.now()
    print("Stop:", timeStop)
    print("Czas:", timeStop - timeStart)

    print("Koniec")

main()
```
